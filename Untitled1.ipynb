Importing the necessary dependencies

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
movieData = pd.read_csv('movie_metadata.csv')
movieData.drop_duplicates()
```

```python
movieData.head(5)
```

```python
list(movieData.columns.values)
```

```python
selectedFeatures=['movie_title','actor_1_name','actor_2_name','actor_3_name','genres','imdb_score','plot_keywords']
for feature in selectedFeatures:
  movieData[feature]=movieData[feature].fillna('')
```

Pruned df

```python
prunedMovieData=movieData['movie_title']
for i in range(1,len(selectedFeatures)):
  prunedMovieData=prunedMovieData+' '+str(movieData[selectedFeatures[i]])

print(prunedMovieData)
```

text to vector

```python
vectorizer = TfidfVectorizer()
featureVectors = vectorizer.fit_transform(prunedMovieData)
print(featureVectors)
```

Getting the cosine similarity

```python
similarity = cosine_similarity(featureVectors)
print(similarity.shape)
```

```python
allTitles = movieData['movie_title'].tolist()
print(allTitles)
```

```python
movie=input("Enter your favorite movie: ")
movie=difflib.get_close_matches(movie,allTitles)
try:
  movie=movie[0]
  print(movie)
except:
  print("no such movie")
```

```python
ind=allTitles.index(movie)
simScore=list(enumerate(similarity[ind]))
simScore=sorted(simScore,key=lambda x:x[1],reverse=True)
print(simScore)
```

```python
recommendedMovies=[]
for i in range(30):
  recommendedMovies.append(allTitles[simScore[i][0]])
print(recommendedMovies)
```
